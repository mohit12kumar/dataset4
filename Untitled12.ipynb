```python
data.drop([
    'CANCELLATION_CODE',
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT',
    'FL_NUMBER',
    'AIRLINE_DOT',
    'DOT_CODE',
    'TAXI_OUT',
    'WHEELS_OFF',
    'WHEELS_ON',
    'TAXI_IN'
], axis=1, inplace=True)
```

```python
data['ELAPSED_TIME'].fillna(data['CRS_ELAPSED_TIME'], inplace=True)
data['AIR_TIME'].fillna(data['ELAPSED_TIME'], inplace=True)
```

```python
data['ELAPSED_TIME'].fillna(data['ELAPSED_TIME'].mean(), inplace=True)
```

```python
data['AIR_TIME'].fillna(data['AIR_TIME'].mean(), inplace=True)
```

```python
data['FL_DATE'] = pd.to_datetime(data['FL_DATE'])

data['month'] = data['FL_DATE'].dt.month
data['day'] = data['FL_DATE'].dt.day
data['day_of_week'] = data['FL_DATE'].dt.dayofweek

# Weekend feature
data['is_weekend'] = data['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```
